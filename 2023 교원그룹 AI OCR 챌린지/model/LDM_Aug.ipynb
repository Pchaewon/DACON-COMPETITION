```python
import torch
from diffusers import LDMSuperResolutionPipeline
```

```python
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
device
```

```python
select1 = random.randint(0, 76888)
imgFile1 = f'./train/TRAIN_{select1:05d}.png'
```

```python
model_id = "CompVis/ldm-super-resolution-4x-openimages"

# load model and scheduler
pipeline = LDMSuperResolutionPipeline.from_pretrained(model_id)
pipeline = pipeline.to(device)
```

```python
from PIL import Image
```

```python
idx = 76888
for i in range(idx+1):
  low_res_img = Image.open(f'train/TRAIN_{i:05d}.png')
  low_res_img = low_res_img.resize((100,125))

  upscaled_image = pipeline(low_res_img, num_inference_steps=100, eta=1).images[0]
  
  new_data = {
    'id': f"TRAIN_{idx:05d}",
    'img_path' : f'./CutMix_LDM/TRAIN_{i:05d}.png',
    'label': df['label'][i],
    'len':df['len'][i] 
  }
  
  df = df.append(new_data, ignore_index=True)
  
  # save image
  upscaled_image.save(f"CutMix_LDM/TRAIN_{i:05d}.png")
  len2count[df['len'][i]] += 1
```
